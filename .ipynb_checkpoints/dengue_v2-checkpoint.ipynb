```python
#load data
dengue_features_train = pd.read_csv("C:\\Users\\user\\Desktop\\python workspace\\drivendata\\dengue_features_train.csv")
dengue_features_test =  pd.read_csv("C:\\Users\\user\\Desktop\\python workspace\\drivendata\\dengue_features_test.csv")
dengue_labels_train = pd.read_csv("C:\\Users\\user\\Desktop\\python workspace\\drivendata\\dengue_labels_train.csv")
```

```python
#merge features and labels of training data
df = pd.merge(dengue_features_train, dengue_labels_train, how='outer', on=dengue_labels_train.columns.tolist()[:-1])
df.head(10)
```

```python
df.shape
```

特徵處理

```python
#see Nan
df.isnull().sum()
```

```python
#將所有遺漏值以中位數方式填入，避免缺失值造成的影響
df['ndvi_ne'] = df['ndvi_ne'].fillna(df['ndvi_ne'].median())
df['ndvi_nw'] = df['ndvi_nw'].fillna(df['ndvi_nw'].median())
df['ndvi_se'] = df['ndvi_se'].fillna(df['ndvi_se'].median())
df['ndvi_sw'] = df['ndvi_sw'].fillna(df['ndvi_sw'].median())
df['precipitation_amt_mm'] = df['precipitation_amt_mm'].fillna(df['precipitation_amt_mm'].median())
df['reanalysis_air_temp_k'] = df['reanalysis_air_temp_k'].fillna(df['reanalysis_air_temp_k'].median())
df['reanalysis_avg_temp_k'] = df['reanalysis_avg_temp_k'].fillna(df['reanalysis_avg_temp_k'].median())
df['reanalysis_dew_point_temp_k'] = df['reanalysis_dew_point_temp_k'].fillna(df['reanalysis_dew_point_temp_k'].median())
df['reanalysis_max_air_temp_k'] = df['reanalysis_max_air_temp_k'].fillna(df['reanalysis_max_air_temp_k'].median())
df['reanalysis_min_air_temp_k'] = df['reanalysis_min_air_temp_k'].fillna(df['reanalysis_min_air_temp_k'].median())
df['reanalysis_precip_amt_kg_per_m2'] = df['reanalysis_precip_amt_kg_per_m2'].fillna(df['reanalysis_precip_amt_kg_per_m2'].median())
df['reanalysis_relative_humidity_percent'] = df['reanalysis_relative_humidity_percent'].fillna(df['reanalysis_relative_humidity_percent'].median())
df['reanalysis_sat_precip_amt_mm'] = df['reanalysis_sat_precip_amt_mm'].fillna(df['reanalysis_sat_precip_amt_mm'].median())
df['reanalysis_specific_humidity_g_per_kg'] = df['reanalysis_specific_humidity_g_per_kg'].fillna(df['reanalysis_specific_humidity_g_per_kg'].median())
df['reanalysis_tdtr_k'] = df['reanalysis_tdtr_k'].fillna(df['reanalysis_tdtr_k'].median())
df['station_avg_temp_c'] = df['station_avg_temp_c'].fillna(df['station_avg_temp_c'].median())
df['station_diur_temp_rng_c'] = df['station_diur_temp_rng_c'].fillna(df['station_diur_temp_rng_c'].median())
df['station_max_temp_c'] = df['station_max_temp_c'].fillna(df['station_max_temp_c'].median())
df['station_min_temp_c'] = df['station_min_temp_c'].fillna(df['station_min_temp_c'].median())
df['station_precip_mm'] = df['station_precip_mm'].fillna(df['station_precip_mm'].median())
```

```python
df.head(10) #檢查填入中位數後的資料樣貌
```

```python
df.isnull().sum() #檢查是否還有遺漏值
```

```python
df.shape
```

```python
#將資料切成train和test資料集
from sklearn.cross_validation import train_test_split

data_train = df.iloc[:,4:24]
data_test = df.iloc[:,24]

train_X, test_X, train_y, test_y = train_test_split(data_train, data_test, test_size = 0.3)
```

```python
#建立線性回歸模型並將資料餵入
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(train_X, train_y)
```

```python
#用訓練好的模型來預測
y_pred = clf.predict(test_X)
y_pred=y_pred.round()
y_pred
```

```python
#以MAE評估模型效能
from sklearn.metrics import mean_absolute_error, r2_score
# The coefficients
print('Coefficients: \n', clf.coef_)

# The mean absolute error
print("Mean absolute error: %.2f"
      % mean_absolute_error(test_y, y_pred))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(test_y, y_pred))
```
